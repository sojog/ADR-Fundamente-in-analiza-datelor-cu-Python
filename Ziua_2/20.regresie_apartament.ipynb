```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# m2 / pret
apartamente = [ (37, 65), (45, 85), (60, 120), (50, 112), (70, 130), (55, 80), (45, 72)]
```

```python
x = [ i[0] for i in apartamente]
y = [ i[1] for i in apartamente]
x, y
```

```python
plt.scatter(x, y)
```

```python
def functie(x):
    return 1.3 * x + 30
y_prezis = [functie(i) for i in x]

plt.scatter(x, y)
plt.plot(x, y_prezis)
```

```python
from scipy import stats


rezultate = stats.linregress(x, y)
rezultate
```

```python
slope, intercept, *others = rezultate
```

```python
def functie(x):
    return slope * x + intercept
y_prezis = [functie(i) for i in x]

plt.scatter(x, y)
plt.plot(x, y_prezis)
```
